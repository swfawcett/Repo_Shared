# Global Power Plant Visualizations with Plotly

Published on Resource Watch and Google Earth Engine. http://resourcewatch.org/ https://earthengine.google.com/ URI: http://datasets.wri.org/dataset/globalpowerplantdatabase

##### Importing the necessary libararies 


```python
import pandas as pd
import plotly.express as px
import os
```

##### Import power plant plant .csv as pandas data frame

```python
df = pd.read_csv("global_power_plant_database.csv")
df
```

##### Group power plants by country, then sum megawatt capacity, and create scatter_geo map denoting overall capacity for each countrie's power plants

```python
sum_mw = df.groupby(['country','country_long'], as_index=False ).capacity_mw.sum()
sum_mw

fig = px.scatter_geo(sum_mw, locations='country', color='capacity_mw', 
                    size='capacity_mw',hover_name='country_long', projection='natural earth',
                    title='Global Megatwatt Capacity per Country', template='plotly_dark',
                    labels={'capacity_mw':'Capacity Megawatt'})
fig.show()
```

#### Pie graph for total global megawatt capaicty by primary fuel type

```python
fig = px.pie(df, names='primary_fuel', values='capacity_mw', template='plotly_dark',
title='Megawatt Capacity per Fuel Type Globally')
fig.show()
```

#### Group by primary fuel and get size to illustrate total plants per type 

```python
fuel_sum = df.groupby(['primary_fuel'],as_index=False).size().sort_values('size',ascending=False)
fuel_sum
fig = px.bar(fuel_sum, x='primary_fuel',y='size',color='primary_fuel',
labels={'size':'# of Plants','primary_fuel':'Primary Fuel'},
template='plotly_dark', title='Total # of Plants per Fuel Type')
fig.show()
```

```python
us_plants = df.query("country=='USA'")
us_plants
fig = px.scatter_map(us_plants,lon='longitude',hover_data=['primary_fuel','gppd_idnr'], template='plotly_dark',
                        lat='latitude',color='primary_fuel',size='capacity_mw', color_continuous_scale='twilight_r',
                        map_style='carto-darkmatter',hover_name='name',title='US Power Plants',height=700,width=1400)
fig.show()
```
